```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import cvlib
import cv2
```

```python
test1 = cv2.imread("images/test1.jpg")
test1 = cv2.cvtColor(test1, cv2.COLOR_BGR2RGB)
test1 = cv2.resize(test1,(224,224))
plt.imshow(test1)
plt.title("test1")
plt.show()
```

```python
test2 = cv2.imread("images/test2.jpg")
test2 = cv2.cvtColor(test2, cv2.COLOR_BGR2RGB)
test2 = cv2.resize(test2,(224,224))
plt.imshow(test2)
plt.title("test2")
```

```python
test3 = cv2.imread("images/test3.jpeg")
test3 = cv2.cvtColor(test3, cv2.COLOR_BGR2RGB)
test3 = cv2.resize(test3,(224,224))
plt.imshow(test3)
plt.title("test3")
```

```python
test4 = cv2.imread("images/test4.jpg")
test4 = cv2.cvtColor(test4, cv2.COLOR_BGR2RGB)
test4 = cv2.resize(test4,(224,224))
plt.imshow(test4)
plt.title("test4")
```

```python
f, axarr = plt.subplots(2,2)
axarr[0,0].imshow(test1) 
axarr[0,1].imshow(test2)
axarr[1,0].imshow(test3)
axarr[1,1].imshow(test4)
plt.show()
```

```python
test1 = np.expand_dims(test1, 0)
test2 = np.expand_dims(test2, 0)
test3 = np.expand_dims(test3, 0)
test4 = np.expand_dims(test4, 0)
```

```python
model = tf.keras.models.load_model("model.h5")
class_name =['WithMask', 'WithoutMask']
```

```python
print(f"Test 1 {class_name[np.argmax(model.predict(test1))]}")
print(f"Test 2 {class_name[np.argmax(model.predict(test2))]}")
print(f"Test 3 {class_name[np.argmax(model.predict(test3))]}")
print(f"Test 4 {class_name[np.argmax(model.predict(test4))]}")
```
