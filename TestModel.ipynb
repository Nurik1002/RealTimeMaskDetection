```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import cvlib
import cv2
```

```python
model = tf.keras.models.load_model("model.h5")
class_name = ['WithMask', 'WithoutMask']
```

```python
test1 = cv2.imread("images/test1.jpg")
test1 = cv2.cvtColor(test1, cv2.COLOR_BGR2RGB)
test1 = cv2.resize(test1,(224,224))
plt.imshow(test1)
plt.title("Test 1 img")
plt.xticks([])
plt.yticks([])
plt.show()
test1 = np.expand_dims(test1, 0)
print(f"Test 1 {class_name[np.argmax(model.predict(test1))]}")
```

```python
test2 = cv2.imread("images/test2.jpg")
test2 = cv2.cvtColor(test2, cv2.COLOR_BGR2RGB)
test2 = cv2.resize(test2,(224,224))
plt.imshow(test2)
plt.title("test2")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
face2 = cvlib.detect_face(test2)
face2 = face2[0][0]
face_2 = test2[face2[0]:face2[2], face2[1]:face2[3]]
face_2 = cv2.resize(face_2, (224,224))
plt.imshow(face_2)
plt.xticks([])
plt.yticks([])
face_2 = np.expand_dims(face_2, 0)
plt.title(class_name[np.argmax(model.predict(face_2))])
plt.show()
```
